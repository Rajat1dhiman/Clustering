1. What is unsupervised learning in the context of machine learning ?

Ans. Unsupervised learnig is a type of machine learning where the model is trained without labeled data. Instead of learning from input-outputs pairs , the algorithm tries to identify patterns, structures or relationships within the input data itself.

**Key Characteristics**

1. **No Labeled Outputs:** The data only contains input features, no corresponding target values.

2. **Goal** : Discover hidden structures or grouping in the data.

2. How does K-Means clustering algorithm work ?

Ans. **Step-by-Step Working of K-Means Algorithm:**

1. **Choose the number of clusters**
ùêæ
K

* This is set manually.

* Example: If K = 3, we want to divide the data into 3 clusters.

2. **Initialize centroids**

* Randomly pick K data points as the initial cluster centers (called centroids).

3. **Assign each data point to the nearest centroid**

* Use Euclidean distance (most common) to assign each point to the cluster whose centroid is closest.

* This forms K clusters.

4. **Recalculate the centroids**

* For each cluster, calculate the mean position of all the points in that cluster.

* This mean becomes the new centroid.

5. **Repeat Steps 3 & 4**

* Reassign points to the nearest centroids and recalculate.

* Continue until:

   * The centroids don‚Äôt change (or),

   *  A maximum number of iterations is reached.

3. Explain the concept of a dendrogram in hierarchical clustering.

Ans. A dendrogram is a tree-like diagram used to visualize the process of hierarchical clustering. It shows how data points or clusters are merged step by step, allowing us to understand the structure and relationships in the data.

Hierarchical clustering builds a hierarchy of clusters either:

* **Agglomeratively** (bottom-up): Start with each point as its own cluster, then merge.

* **Divisively (top-down)**: Start with one cluster and split recursively.

Most commonly used is **agglomerative clustering.**



4. What is the main difference between K-Means and Hierarchical Clustering ?

Ans. **Use K-Means if:**

* You know the number of clusters.

* You have a large dataset.

* You want speed and scalability.

**Use Hierarchical Clustering if:**

* You don‚Äôt know how many clusters you want.

* You want to visualize the data‚Äôs cluster structure (via a dendrogram).

* You‚Äôre working with a small to medium-sized dataset.



5. What are the advantages of DBSCAN over K-Means ?

Ans. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several key advantages over K-Means, especially when dealing with irregular cluster shapes and noisy data.

* DBSCAN automatically determines the number of clusters based on density.

* Explicitly identifies noise points as outliers.

* Can find clusters of various shapes (e.g., crescent, spiral).

* Works with density and distance, not centroids

8. Why is feature scaling important in clustering algorithm like K-Means ?

Ans.  **Why Feature Scaling Matters in K-Means:**

1. **K-Means Uses Euclidean Distance**

* K-Means clusters data by minimizing the sum of squared distances from each point to the cluster centroid.

* Euclidean distance is sensitive to the scale of features.

**For example:**

If you have Age (0‚Äì100) and Income (0‚Äì100,000), the distance will be dominated by Income unless the data is scaled.

2. **Unscaled Features Lead to Biased Clustering**

* Features with larger ranges will dominate the distance metric, effectively ignoring other important features.

* This can result in:

  * Incorrect cluster assignment

  * Wrong number of clusters

  * Poor performance or interpretability


9. How does DBSCAN identify noise points ?

Ans. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points by checking whether a point does not belong to any dense region of the dataset.

**Classification of Points in DBSCAN:**

1. **Core Point** :	Has at least min_samples within eps radius (including itself).

2. **Border Point** :	Has fewer than min_samples within eps, but is within the eps neighborhood of a core point.

3. **Noise Point (Outlier):**	Not a core point and not within eps of any core point. These are not assigned to any cluster.

11. What is the elbow method in K-Means clustering ?

Ans. The **Elbow Method** is a popular technique used in K-Means clustering to help determine the **optimal number of clusters (K).**

**Goal:**

Find the number of clusters
ùêæ
K where adding more clusters doesn‚Äôt significantly improve the clustering result ‚Äî i.e., where the inertia (within-cluster sum of squares) starts to level off.

**How It Works:**

1. Run K-Means for a range of
ùêæ
K values (e.g., from 1 to 10).

2. For each value of
ùêæ
K, compute the inertia (i.e., how compact the clusters are).

3. Plot
ùêæ
K vs. inertia.

4. Look for the point where the inertia decreases sharply up to a point, then slows down.
This point is called the ‚Äúelbow‚Äù, and it indicates the optimal K.



12. Describe the concept of "density" in DBSCAN.

Ans. In **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**, the concept of "density" is fundamental. It refers to how closely data points are packed together in a region ‚Äî **dense areas form clusters, while sparse areas are treated as noise.**

**How DBSCAN Defines Density:**

DBSCAN uses two parameters to measure density:

* eps (epsilon)
‚Üí The radius that defines the neighborhood around a point.

* min_samples
‚Üí The minimum number of points (including the point itself) required to form a dense region.




13. Can hierarchical clustering be used on categorical data?

Ans. Yes, **Hierarchical Clustering can be used on categorical data**, but it requires modifications in how distances are calculated, since traditional distance measures like Euclidean distance are not suitable for non-numeric (categorical) features.

**Why It's Tricky:**

* Hierarchical clustering depends heavily on distance or similarity metrics.

* Categorical data (like "red", "blue", "green" or "low", "medium", "high") cannot be compared using numerical distances without preprocessing.

15. Explain the term "linkage criteria" in hierarchical clustering.

Ans. In hierarchical clustering, the term **"linkage criteria"** (or linkage method) defines how the distance between two clusters is calculated when merging them. It directly affects the structure of the resulting **dendrogram** and the final clusters.

**Why Linkage Criteria Matter:**

During agglomerative clustering (bottom-up approach), at each step, the algorithm merges the **two closest clusters**. The linkage method determines what ‚Äúclosest‚Äù means when comparing clusters that contain more than one point.

16.  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

Ans. **K-Means clustering can perform poorly on data with varying cluster sizes or densities** because it makes strong assumptions about the structure of the clusters. Specifically, K-Means assumes that:

**Key Assumptions of K-Means:**

1. Clusters are spherical and equally shaped (convex)

2. All clusters have roughly the same size

3. All clusters have similar densities

4. Uses Euclidean distance, which is sensitive to scale and density

**Example:**
Imagine this dataset:

* One cluster is large and spread out.

* One is small but tightly packed.

* One is irregularly shaped.

**K-Means may:**

* Assign distant points to the wrong cluster

* Split a large cluster into two

* Merge two small, nearby clusters into one



17. What are the core parameters in DBSCAN , and how do they influence clustering ?

Ans. In **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**, the clustering behavior is controlled by two core parameters:

1. **eps (Epsilon Radius)**

* The maximum distance between two points for them to be considered as neighbors.

**Effect:**

* Small eps: Fewer neighbors ‚Üí more points marked as noise, possibly many small clusters

* Large eps: More neighbors ‚Üí fewer clusters, possibly merging different clusters into one

2. **min_samples (Minimum Samples)**

* The minimum number of points (including the point itself) required to form a dense region (i.e., a core point).

**Effect:**

* Small min_samples (e.g., 2 or 3): Clusters form easily, including small or noisy groups

* Large min_samples: Only denser areas are accepted as clusters ‚Üí more points labeled as noise

19. What is agglomerative clustering ?

Ans. **Agglomerative Clustering is a type of Hierarchical Clustering** that builds clusters using a bottom-up approach ‚Äî starting with each data point as its own cluster and gradually merging the closest clusters until only one or a few clusters remain.

**Key Concept:**

Agglomerative = ‚ÄúTo gather together‚Äù
It merges smaller clusters into larger ones step by step based on a distance or similarity measure.

20. What makes Silhoutte Score a better metric than just inertia for model evaluation ?

Ans. **Why Silhouette Score Is Better:**

1. **Accounts for both cohesion and separation**

* Cohesion: How close points are to their own cluster

* Separation: How far they are from the next nearest cluster

High silhouette score = tight and well-separated clusters

2. **Normalized and interpretable**

* Range: -1 to +1

  * +1 ‚Üí Well-clustered

  *   0 ‚Üí Borderline

  * Negative ‚Üí Probably misclassified

* Easy to compare across different values of K or algorithms

3. **Doesn‚Äôt always prefer more clusters**

* In contrast to inertia (which **always decreases** with increasing K), silhouette score helps **find the optimal number of clusters**, making it ideal for the Elbow + Silhouette method combo.

4. **Model-agnostic**

* Can be used with any clustering algorithm, not just K-Means

* Works even when clusters are non-spherical or vary in size

* Example: Elbow vs. Silhouette

Suppose you run K-Means for K = 2 to 10:

Inertia keeps dropping ‚Üí unclear where to stop

Silhouette score peaks at K = 3 ‚Üí clear signal of optimal clustering

21. Generate synthetic data with 4 centres using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# 2. Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 3. Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 1. Generate synthetic "moons" data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# 2. Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# 3. Plot clusters and highlight outliers
plt.figure(figsize=(8, 6))

# Identify core points, border points, and noise
unique_labels = set(labels)
colors = ['blue', 'green', 'orange', 'purple', 'cyan']

for label in unique_labels:
    if label == -1:
        # Outliers
        plt.scatter(X[labels == label, 0], X[labels == label, 1],
                    c='red', label='Outliers', marker='x', s=50)
    else:
        plt.scatter(X[labels == label, 0], X[labels == label, 1],
                    c=colors[label % len(colors)], label=f'Cluster {label}')

plt.title("DBSCAN Clustering on make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k=2 to 5. displays as a bar chart.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Generate synthetic data
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# 2. Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# 3. Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for KMeans Clustering")
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

30. Load the Iris dataset and use hierarchical clustering to group data.Plot a dendrogram with average linkage.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data

# 2. Perform hierarchical clustering with 'average' linkage
linked = linkage(X, method='average')

# 3. Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=False)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# 2. Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_

# 3. Create a mesh grid for decision boundary plotting
h = 0.1  # step size in mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster labels for each point in the mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 4. Plot decision boundaries and clusters
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Accent')  # decision boundaries
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='Accent', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Blobs)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

44. Load the Iris dataset and plot the Silhoutte Coefficient for each sample after KMeans clustering.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Apply KMeans clustering
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(X)

# Step 3: Compute silhouette scores
silhouette_vals = silhouette_samples(X, cluster_labels)
avg_score = silhouette_score(X, cluster_labels)

# Step 4: Plot silhouette scores
fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10

for i in range(n_clusters):
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / n_clusters)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_vals,
                     facecolor=color, edgecolor=color, alpha=0.7)

    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # space between clusters

# Draw average score line
ax.axvline(x=avg_score, color="red", linestyle="--")
ax.set_title("Silhouette Plot for KMeans on Iris Dataset")
ax.set_xlabel("Silhouette Coefficient")
ax.set_ylabel("Cluster")
plt.grid(True)
plt.show()
```
